def min_max_scaling(data_list):
    minimum_value = min(data_list)
    maximum_value = max(data_list)
    scaled_data_list = [(((x - minimum_value) / (maximum_value - minimum_value))*(1-0)) for x in data_list]
    return scaled_data_list
