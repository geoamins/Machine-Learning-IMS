### Linear Regression using Least Squares method
In statistics, linear regression is a linear approach to modelling the relationship between a dependent variable and one or more independent variables. 

Let X be the independent variable and Y be the dependent variable. We will define a linear relationship between these two variables as follows: 
$ y = mX+c $

* m is the slope of the line, and 
* c is the y intercept

**Our challenege today is to determine the value of m and c, that gives the minimum error for the given dataset. We will be doing this by using the Least Squares method.**

### Least Squares method
In order to minimize the loss function value, this is done using partial derivative of L, equating it to 0 and then finding an expression for the m and c.

Here is the mean of all the values in the input X and is the mean of all the values in the desired output Y. This is the Least Squares method. 

```python
# Making imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (8.0, 4.0)   # width and height
```

```python
# Preprocessing Input data
data = pd.read_csv('data.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
plt.scatter(X, Y)
plt.show()
```

```python
# Building the model
X_mean = np.mean(X)
Y_mean = np.mean(Y)

num = 0
den = 0
for i in range(len(X)):
    num += (X[i] - X_mean)*(Y[i] - Y_mean)
    den += (X[i] - X_mean)**2
m = num / den
c = Y_mean - m*X_mean

print (m, c)
```

```python
# Making predictions
Y_pred = m*X + c

plt.scatter(X, Y) # actual
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.show()
```
