### Logistic Regression From Scratch (1 Feature)

```python
import numpy as np

# Dataset
X = np.array([1, 2, 3]).reshape(-1, 1)   # feature
y = np.array([0, 0, 1]).reshape(-1, 1)   # labels
```

```python
# Initialize parameters
w = np.zeros((1, 1))  # w=[0​]
b = 0
alpha = 0.1
n = len(y)
```

```python
# Sigmoid function

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Training loop

for i in range(5):
    
    # Forward pass
    z = np.dot(X, w) + b
    y_hat = sigmoid(z)
    
    # Compute loss (Binary Cross Entropy)
    loss = -(1/n) * np.sum(
        y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat)
    )
    
    # Compute gradients
    dw = (1/n) * np.dot(X.T, (y_hat - y))
    db = (1/n) * np.sum(y_hat - y)
    
    # Update parameters
    w = w - alpha * dw
    b = b - alpha * db
    
    # Print results
    print(f"Iteration {i+1}")
    print(f"w = {w.flatten()[0]:.6f}, b = {b:.6f}")
    print(f"Loss = {loss:.6f}")
    print("-------------------------")
```
