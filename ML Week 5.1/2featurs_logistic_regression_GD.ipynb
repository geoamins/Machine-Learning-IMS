```python
# Logistic Regression with 2 Features (from scratch) + Decision Boundary
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Dataset (2 Features)
# -----------------------------
# Example: [Study Hours, Sleep Hours]
X = np.array([
    [1, 2],
    [2, 1],
    [3, 3],
    [4, 5],
    [5, 4]
])

y = np.array([0, 0, 0, 1, 1]).reshape(-1, 1)
```

```python
# Initialize parameters
# -----------------------------
m, n = X.shape
w = np.zeros((n, 1))
b = 0
alpha = 0.1
```

```python
# Sigmoid function
# -----------------------------
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Training loop
# -----------------------------
iterations = 500

for i in range(iterations):
    z = np.dot(X, w) + b
    y_hat = sigmoid(z)
        
    loss = -(1/m) * np.sum(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
    
    dw = (1/m) * np.dot(X.T, (y_hat - y))
    db = (1/m) * np.sum(y_hat - y)
    
    w = w - alpha * dw
    b = b - alpha * db
```

```python
print("Final Weights:\n", w)
print("Final Bias:\n", b)
print("Final Loss:", loss)
```
