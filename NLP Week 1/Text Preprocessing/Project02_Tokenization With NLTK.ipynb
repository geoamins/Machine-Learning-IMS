# Import Libraries

```python
import nltk
from nltk import word_tokenize
```

# NLTK Language Model

```python
nltk.download('punkt') # It is pretrained model for tokenization
```

# Perform Tokenization

```python
sentence = '"Today is Friday. Next Week, Were coming from U.S.!\"'
```

```python
for token in word_tokenize(sentence):
  print(token)
```

```python
sentence.split()
```

```python
sent2 = "We're here to help you! ask the question, email support@abcnet.ax or visit us at http://www.abcnet.ax!"
```

```python
for token in word_tokenize(sent2):
  print(token)
```

```python
sent3 = "We are learning NLP. This is a good course. Is weather good today?"
```

```python
for token in word_tokenize(sent3):
  print(token)
```

```python
sent3.split()
```
